# 📊 Task 1: Data Aggregation & Analysis

**Objective:** Analyze the supermarket transaction dataset to extract actionable business insights for InsightSpark.

**Business Questions to Answer:**
1. Across locations, how many apples were purchased in cash?
2. How much total cash was spent on these apples?
3. Across all payment methods, how much money was spent at the Bakershire store location by non-member customers?

**Methodology:**
1. **Load & Profile:** Ingest data and understand the schema.
2. **Validate:** Ensure data types and critical columns are clean.
3. **Aggregate:** Use Pandas filtering and grouping to answer the questions.
4. **Deliver:** Export findings to `deliverables/task1.xlsx`.

```python
import numpy as np
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/beingAnujChaudhary/cba-insightspark-data-platform/main/data/raw/supermarket_transactions.xlsx"

transactionData = pd.read_excel(url)
```

```python
# Show first 5 rows

transactionData.head()
```

## 🔍 Step 2: Data Profiling & Validation
Before answering questions, we must understand the structure of our data. We need to verify column names, data types, and check for any missing values that could skew our financial calculations.

```python
# WHAT WE ARE DOING: Checking the shape, columns, and data types.
# WHY: To ensure 'total_amount' and 'quantity' are numeric, and 'timestamp' is a datetime object.
print(f"📊 Dataset Shape: {transactionData.shape}")
print("\n📋 Column Names:")
print(transactionData.columns.tolist())
```

```python
print("\n📊 Data Types & Missing Values:")
# .info() gives us memory usage, data types, and non-null counts
print(transactionData.info())
```

```python
# WHAT WE ARE DOING: Checking for unique values in categorical columns we will filter on.
# WHY: To ensure 'apple' is exactly spelled 'apple' (and not 'Apples' or 'apple ') 
# and 'cash' is exactly 'cash'.
print("\n🔍 Unique Payment Methods:")
print(transactionData['payment_method'].unique())
```

```python
print("\n🔍 Unique Customer Types:")
print(transactionData['customer_type'].unique())
```

```python
transactionData["store"].unique()
```

```python
transactionData["product_name"].unique()
```

```python
# The goal is to isolate all variations of 'apple' in the 'product_name' column.
# We use .str.contains() because users might type 'Apple', 'apple', or 'Apples'.

print("\n🔍 Products containing 'apple':")

# 1. transactionData["product_name"].str.contains("apple", case=False, na=False)
#    - .str.contains("apple"): Looks for the string "apple" inside product names.
#    - case=False: Makes the search case-insensitive (finds "Apple" and "apple").
#    - na=False: If a cell is empty (NaN), treat it as False (don't include it in results).

# 2. [...] : This uses the boolean mask above to filter the entire dataframe.

apple_products = transactionData[
    transactionData["product_name"].str.contains("apple", case=False, na=False)
]["product_name"].unique() 

# 3. ["product_name"].unique():
#    - We select only the 'product_name' column from our filtered list.
#    - .unique(): Returns only one instance of each distinct apple product 
#      (e.g., if 'Granny Smith Apple' appears 100 times, it only lists it once).

# FIX: Added () after .unique because .unique is a function that must be called.

print(apple_products)
```

## 🍎 Question 1: Apples Purchased in Cash (By Location)
**Question:** *Across locations, how many apples were purchased in cash?*

**Approach:** 
1. Filter the dataset for `product_name == 'apple'` (exact match to exclude juice/sauce).
2. Filter for `payment_method == 'cash'`.
3. Group by `store` and sum the `quantity` column.

```python
# WHAT WE ARE DOING: Creating boolean masks for filtering.
# WHY: Using masks is highly readable and allows us to reuse conditions if needed.
is_apple = transactionData["product_name"].str.lower() == "apple"
is_cash = transactionData["payment_method"].str.lower() == "cash"
```

```python
# Apply filters
apples_cash_df = transactionData[is_apple & is_cash]
```

```python
# WHAT WE ARE DOING: Grouping by store and summing the quantity.
# WHY: The question asks "how many apples", which implies the total count of items (quantity), 
# not just the number of transactions (rows).
q1_result = apples_cash_df.groupby('store')['quantity'].sum().reset_index()
q1_result.columns = ['Store_Location', 'Total_Apples_Purchased_Cash']
```

```python
# Sort by highest volume for better readability
q1_result = q1_result.sort_values(by='Total_Apples_Purchased_Cash', ascending=False)
```

```python
print("✅ QUESTION 1: Apples purchased in cash by location")
display(q1_result)
```

```python
# WHAT WE ARE DOING: Summing the 'total_amount' column for our filtered dataframe.
# WHY: This gives us the exact revenue generated from cash purchases of apples.
q2_result = apples_cash_df['total_amount'].sum()

print("✅ QUESTION 2: Total cash spent on apples")
print(f"💰 Total Revenue: ${q2_result:,.2f}")
```

## 🏬 Question 3: Bakershire Non-Member Spend
**Question:** *Across all payment methods, how much money was spent at the Bakershire store location by non-member customers?*

**Approach:**
1. Filter for `store == 'Bakershire'`.
2. Filter for `customer_type == 'non-member'`.
3. Sum the `total_amount`.

```python
# WHAT WE ARE DOING: Filtering for the specific store and customer tier.
# WHY: InsightSpark wants to understand the revenue contribution of non-loyalty customers at this specific branch.
is_bakershire = transactionData['store'].str.lower() == 'bakershire'
is_non_member = transactionData["customer_type"].str.lower() == "non-member"

print(is_bakershire.sum(), is_non_member.sum())
```

```python
bakershire_non_member_df = transactionData[is_bakershire & is_non_member]
```

```python
# Calculate total spend
q3_result = bakershire_non_member_df['total_amount'].sum()
```

```python
# Bonus: Let's also see how many transactions this represents
transaction_count = len(bakershire_non_member_df)

print("✅ QUESTION 3: Bakershire Non-Member Spend")
print(f"💰 Total Spend: ${q3_result:,.2f}")
print(f"📝 Number of Transactions: {transaction_count}")
```

## 📤 Step 6: Exporting Deliverables
As per our project architecture, we must save our findings to the `deliverables/` folder. We will export the aggregated data to an Excel file with multiple sheets for the stakeholders.

```python
import os

# WHAT WE ARE DOING: Creating the deliverables directory if it doesn't exist.
# WHY: Ensures the code doesn't crash if the folder hasn't been created manually yet.
os.makedirs('deliverables', exist_ok=True)

# Create a summary dataframe for Q2 and Q3 since they are single values
summary_data = {
    'Metric': ['Total Cash Spent on Apples', 'Bakershire Non-Member Spend'],
    'Value': [f"${q2_result:,.2f}", f"${q3_result:,.2f}"]
}
summary_df = pd.DataFrame(summary_data)
```

```python
# WHAT WE ARE DOING: Writing multiple dataframes to a single Excel file.
# WHY: Stakeholders prefer a single Excel file with organized tabs over multiple CSVs.
with pd.ExcelWriter('deliverables/task1_analysis_results.xlsx', engine='openpyxl') as writer:
    q1_result.to_excel(writer, sheet_name='Q1_Apples_by_Location', index=False)
    summary_df.to_excel(writer, sheet_name='Q2_Q3_Financial_Summary', index=False)
    
    # Let's also export the raw filtered data for their own verification
    apples_cash_df.to_excel(writer, sheet_name='Raw_Apples_Cash_Data', index=False)
    bakershire_non_member_df.to_excel(writer, sheet_name='Raw_Bakershire_NonMember', index=False)

print("✅ Deliverables successfully exported to 'deliverables/task1_analysis_results.xlsx'")
```
